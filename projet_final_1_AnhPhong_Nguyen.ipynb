```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import uuid

import xlwings as xw

# lecture/écriture d'un fichier Excel avec xlwings
def save_with_xlwings(file):
    tempfile = f'./{uuid.uuid1()}.xlsx'
    excel_app = xw.App(visible=False)
    excel_book = excel_app.books.open(file)
    excel_book.save(tempfile)
    excel_book.close()
    excel_app.quit()
    return tempfile
```

```python
url = "https://www.vendeeglobe.org/fr/classement"
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
list = soup.find_all(class_="form__input m--select onsubmit-rank")
items = []
filenames = []
for i in list:
    tmp = i.find_all("option")
for item in tmp[1:]:
    date_time = item.text.strip().split()
    date = date_time[1]
    time = date_time[2]
    items.append(item.text.strip().split())
    # print(date,time)
    filenames.append(date[0:4]+date[5:7]+date[8:10]+"_"+time[0:2]+time[3:5]+time[6:8])
    
filenames[:5]
```

```python
# pd.read_excel(f"vendeeglobe_{filenames}.xlsx")
```

```python
def read_excel_concat():
    dfs = [(pd.read_excel(save_with_xlwings(f"vendeeglobe_{filename}.xlsx"))
            .assign(filename=filename))
           for filename in filenames]
    df = (pd.concat(dfs, ignore_index=True))
    return df
read_excel_concat()
```
